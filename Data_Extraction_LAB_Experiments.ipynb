```python
import os
import lyricsgenius 
from dotenv import load_dotenv
import pandas as pd
import time
import json
```

```python
load_dotenv()

CLIENT_ID = os.getenv("client_id")
CLIENT_SECRET = os.getenv("client_secret")
CLIENT_ACCESS_TOKEN = os.getenv("client_access_token")
```

```python
genius = lyricsgenius.Genius(CLIENT_ACCESS_TOKEN)
```

```python
artists = pd.read_csv("./data/100_artists_pop_english.csv")
```

```python
all_artists = set(artists["Artist"])
downloaded_artists = set()
remaining_artists = all_artists - downloaded_artists
```

```python
def get_songs(artist_name, n_songs=10, t_sleep=30):
    while True:
        try:
            artist = genius.search_artist(artist_name, max_songs=n_songs)
            songs = dict()
            for song in artist.songs:
                songs[song.title] = song.lyrics
            return songs
        except Exception as f:
            print(f)
            time.sleep(t_sleep)
```

```python
artists_songs = dict()
for artist_name in artists["Artist"]:
    if artist_name in remaining_artists:
        songs = get_songs(artist_name)
        with open(f"data/jsons/{artist_name.replace(' ', '_')}.json", 'w') as f:
            json.dump(songs, f)
        remaining_artists.remove(artist_name)
```

```python
data = []
for artist_name, artist in artists_songs.items():
    for song in artist.songs:
        data.append((artist_name, song.title, song.lyrics))
```

```python
df = pd.DataFrame(data, columns=("Artist", "Song", "Lyrics")).to_csv("./data/songs.csv")
```
